# Data preparation

```python
df['Bedtime'] = pd.to_datetime(df['Bedtime'])
df['Wakeup time'] = pd.to_datetime(df['Wakeup time'])

#'Female' as 0 and 'Male' as 1
df['Gender'] = df['Gender'].replace({'Female': 0, 'Male': 1})
#'Yes' as 1 and 'No' as 0
df['Smoking status'] = df['Smoking status'].replace({'Yes':1,'No':0})
df.drop(['ID'], axis=1, inplace=True)

df
```

# Target variable and input variables

```python
df['sleep'] = df[['Normalized sleep duration', 'REM sleep percentage', 'Light sleep percentage', 'Deep sleep percentage', 'Normalized Sleep efficiency']].mean(axis=1)

df
```

```python
# Normalize "Sleep duration" based on min and max values
min_sleep_duration = df['Sleep duration'].min()
max_sleep_duration = df['Sleep duration'].max()
df['Normalized sleep duration'] = (df['Sleep duration'] - min_sleep_duration) / (max_sleep_duration - min_sleep_duration) * 100

df['Normalized Sleep efficiency'] = df['Sleep efficiency'] * 100
```

```python
y = df[['sleep']]
y
```

```python
x1 = df[['Age', 'Gender' , 'Smoking status', 'Caffeine consumption', 'Alcohol consumption', 'Exercise frequency', 'Bedtime_hour', 'Wakeup_hour']]
       
x1
```

# Trainset and Testset

```python
#split train set and test set 
from sklearn.model_selection import train_test_split
x1_train, x1_test, y_train, y_test = train_test_split(x1, y, test_size = 0.2, random_state=123)
```

```python
print(x1_train.shape, x1_test.shape, y_train.shape, y_test.shape)
```

# Random Forest

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# find hyperparameters 
para_grid = {
    'n_estimators': [100, 200, 300, 400, 500, 600, 700],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt']
}

# Initialize Random Forest Regressor
rf = RandomForestRegressor()

# Perform Grid Search with cross-validation
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=-1, scoring='neg_mean_squared_error')
grid_search.fit(x1_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators = 100, random_state=123)
rf.fit(x1_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

rf_pred = rf.predict(x1_test)

rf_r2 = r2_score(y_test,rf_pred)
print(f"Random Forest R Sq: {rf_r2}")
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_pred)) 
print(f"Random Forest RMSE: {rf_rmse}")
rf_mae = mean_absolute_error(y_test, rf_pred)
print(f"Random Forest MAE: {rf_mae}")
```

```python
# Find feature importance 
feature_list = list(x1.columns)
feature_imp = pd.Series(rf.feature_importances_, index=feature_list).sort_values(ascending = False)
print("Feature Importance Scores:")
feature_imp
```

# Neural Network 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import regularizers

nn_model = Sequential([
    Dense(128, activation='relu', input_shape=(x1_train.shape[1],), kernel_regularizer=regularizers.l1(0.01)),
    Dense(64, activation='relu', kernel_regularizer=regularizers.l1(0.01)),
    Dense(32, activation='relu', kernel_regularizer=regularizers.l1(0.01)),
    Dense(1, kernel_regularizer=regularizers.l1(0.01))
])

nn_model.compile(optimizer='adam', loss='mean_squared_error')

# Train Neural Network model
nn_model.fit(x1_train, y_train, epochs=200, batch_size=32, validation_split=0.2, verbose=1)
```

```python
# Evaluate Neural Network model using test data
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

nn_pred = nn_model.predict(x1_test).flatten()


nn_r2 = r2_score(y_test, nn_pred)
nn_rmse = np.sqrt(mean_squared_error(y_test, nn_pred))
nn_mae = mean_absolute_error(y_test, nn_pred)
print(f"Neural Network R-squared: {nn_r2}")
print(f"Neural Network RMSE: {nn_rmse}")
print(f"Neural Network MAE: {nn_mae}")
```

```python
from sklearn.inspection import permutation_importance

# permutation importance
results = permutation_importance(nn_model, x1_test, y_test, scoring='neg_mean_squared_error', n_repeats=10, random_state=123)

# Get feature importance scores
importance_scores = results.importances_mean

# Get feature names
feature_names = x1_test.columns  

sorted_indices = importance_scores.argsort()[::-1]

# Print feature importance scores in descending order
print("Feature Importance Scores:")
for idx in sorted_indices:
    print(f"{feature_names[idx]}: {importance_scores[idx]}")

```
